```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/bogus1.xlsx"
```

```python
df= pd.read_excel(path)
```

```python
srednie = df.groupby("schemat")["pacjent 1"].mean()
srednie
```

```python
liczbaist = df.loc[df["pacjent 1"]>3].groupby("schemat")["pacjent 1"].count()
liczbaist
```

```python
fig = plt.figure(figsize = (15, 5))
plt.plot(liczbaist)
```
